```
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_url = 'https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv'

dataframe = pd.read_csv(data_url)
print(dataframe.head(10))
```

```
women = dataframe.Women
men = dataframe.Men

plt.hist(women, bins=15, normed=True)
```

```
fig, ax = plt.subplots()
ax.hist(women, normed=True, histtype='stepfilled', alpha=0.7, label='Women', color='red')
ax.hist(men, normed=True, histtype='stepfilled', alpha=0.7, label='Men', color='blue')
l = ax.legend(prop={'size':10})
for text in l.get_texts():
  text.set_color('white')

ax.set_ylabel('No. of schools')
ax.set_xlabel('Salary (1000s)')

ax.set_facecolor('black')
```

```
schools = dataframe.School

fig = plt.figure(figsize=(10,2))
plt.tight_layout()
ax = fig.add_subplot(111)

bar_width = 0.35
index = np.arange(len(schools))
                
women_bar = ax.bar(index, women, bar_width, label='Women', color='red')
men_bar = ax.bar(index+bar_width, men, bar_width, label='Men', color='blue')
l = ax.legend(prop={'size':10})

ax.set_xticks(index+bar_width/2)
ax.set_xticklabels(schools, rotation=45)

ax.set_xlabel('Schools')
ax.set_ylabel('Salary (1000s)')
ax.set_title('Salaries split by gender from each university')
# fig.xticks(schools, rotation=45)
```
